```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt
```

```python
years = 5
freq = 12
nper = years * freq
pv = 120000
rate_apartment = 0.05
rate_deposit = 0.12
rate_apartment/=freq
rate_deposit/=freq
periods = np.arange(1,nper+1,dtype=int)
```

```python
fv_apartment = - np.around(npf.fv(rate_apartment,periods,0,pv),2)
fv_apartment[-1]
```

Orientacyjna cena mieszkania za 5 lat : 154003.04

```python
monthly_deposit = - np.around(npf.ppmt(rate_deposit,periods,nper,fv_apartment[-1]),2)
np.around(monthly_deposit.sum(),2)
monthly_deposit[0]
```

Aby przy przedstawionej ofercie uzbierać na mieszkanie w ciągu 5 lat, musisz wpłacać do banku każdego miesiąca 1885.68 ?

```python
plt.plot(fv_apartment,label='Mieszkanie')
plt.plot(monthly_deposit.cumsum(),label='Lokata')
plt.title("Miesięczna wartość lokaty oraz mieszkania w ciągu 5 lat")
plt.legend()
plt.xlabel('Liczba okresów')
plt.ylabel('Skumulowana wartość')
plt.grid(True)
```
